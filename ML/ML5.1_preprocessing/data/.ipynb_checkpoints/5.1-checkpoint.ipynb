# Поработайте со столбцом price. Постройте гистограмму, найдите выбросы (ящиковая диаграмма), удалите аномальные значения, постарайтесь привести к нормальному распределению.

```python
cars.boxplot("price");
```

```python
q1 = cars.price.quantile(0.25)
q3 = cars.price.quantile(0.75)
rng = q3-q1
down_d = q1 - 1.5*rng
up_b = q3 + 1.5*rng
```

```python
cars2 = cars[(cars.price<=up_b) * (down_d<=cars.price)]
```

```python
cars2.boxplot("price");
```

```python
cars2.price.hist();
```

# Аналогично поработайте и с другими «важными для целевой функции price» параметрами (powerPS, yearOfRegistration, kilometer и т.д). «Важность» докажите через коэффициент корреляции. Сделайте визуализацию.

```python
%matplotlib inline
for col in ["price", "yearOfRegistration", "powerPS", "kilometer", "monthOfRegistration", "postalCode"]:
    q1 = cars[col].quantile(0.25)
    q3 = cars[col].quantile(0.75)
    rng = q3-q1
    down_d = q1 - 1.5*rng
    up_b = q3 + 1.5*rng
    cars2 = cars2[(cars2[col]<=up_b) * (down_d<=cars2[col])]
    plt.boxplot(cars2[col])
    plt.title(col)
    plt.show()
    plt.hist(cars2[col])
    plt.title(col)
    plt.show()
#     cars2.boxplot(col);
#     cars2[col].hist();
```
