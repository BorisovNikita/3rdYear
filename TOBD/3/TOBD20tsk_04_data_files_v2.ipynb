4) Написать функцию, которая формирует новую запись в телефонной книге, запрашивая у пользователя поочередно все необходимые поля.

```python
def valid(text: str, mask: str):
    while True:
        input_for_valid = input(text)
        if re.search('^'+mask+'$', input_for_valid):
            return input_for_valid

def add_item(ab):
    item = {}
    item['name'] = valid('Введите имя: ', '(\w+\s?)+')
    item['email'] = valid('Введите email: ', '\w+@\w+[.]\w+')
    item['birthday'] = valid(
        'Введите дату рождения (13.03.1999): ',
        '\d{2}[.]\d{2}[.]\d{4}'
    )
    phone_count = int(valid(
        'Введите количество добавляемых телефонов: ',
        '\d+'
    ))
    item['phones'] = []
    for _ in range(phone_count):
        phone = {}
        phone['type'] = valid(
            'Введите тип телефона (home/work): ',
            '(?:home|work)'
        )
        if phone['type'] == 'work':
            phone['phone'] = valid(
                    'Введите номер телефона (111-11-11): ',
                    '\d{3}(?:-\d{2}){2}'
                )
        else:
            phone['phone'] = valid(
                    'Введите номер телефона (+7 (495) 111-11-11): ',
                    '[+]\d [(]\d{3}[)] \d{3}(?:-\d{2}){2}'
                )
                
        item['phones'].append(phone)
    item['id'] = ab[-1]['id']+1
    item['friends'] = []
    yes_no = input('Добавить друга? (yes/no)')
    if yes_no == 'yes':
        friends_count = int(valid(
            'Сколько друзей хотите добавить? ',
            '\d+'
        ))
        
        for _ in range(friends_count):
            item['friends'].append(
                int(valid(
                    'Введите id друга: ',
                    '\d+'
                ))
            )        
    ab.append(item)
    return ab

add_item(my_ab2)
```

```python
from bs4 import BeautifulSoup
with open('.\\addres-book.xml') as f:
    ab = BeautifulSoup(f, 'xml')

res = list();
for person in ab.address_book.find_all("address"):
    ph = [phones.next for phones in person.phones.find_all("phone")]
    res.append({person.find("name").next: ph})
res
```

## Сбор данных с html страниц

```python
import urllib
import urllib.parse
import urllib.request
```

```python
url_wikipedia = 'https://ru.wikipedia.org'
url_es = 'https://ru.wikipedia.org/wiki/Категория:Электростанции_по_алфавиту'
url_es_q  = urllib.parse.quote(url_es, safe='/:')
print(url_es_q)
```

```python
es_bs = BeautifulSoup(urllib.request.urlopen(url_es_q), 'lxml')
```

```python
all_es = []
for i in range(30):
    # load links:
    bef = len(all_es)
#     print(bef)
    for links in es_bs.find_all('div', class_='mw-category-group'):
        for l in links.find_all('a'):
            all_es.append((l['title'], url_wikipedia+l['href']))
```

```python
all_es[0], all_es[200]
```

```python
all_es_prop = []
```

```python
for i, (n, lnk) in enumerate(all_es[:10]):
    es = BeautifulSoup(urllib.request.urlopen(lnk), 'lxml')
    coord = es.find_all('a', class_='mw-kartographer-maplink')
    if len(coord) > 0:
        lat, lon = coord[0]['data-lat'], coord[0]['data-lon']
    else:
        print('{}: {}, COORDINATES IS NOT FOUND! link: {}'.format(i, n, lnk))        
        continue
                          
    es_pr = {}
    table = es.find('table', class_="infobox")
    if table is not None:
        for row in table.find_all('tr'):
            th = row.find_all('th')
            if th and len(th):
    #             print(th[0].text.strip(), end=' || ')
                td = row.find_all('td')
                if td and len(td):
                    es_pr[th[0].text.strip()] = td[0].text.strip()
        #             print(td[0].text.strip())            
        all_es_prop.append((n, lnk, lat, lon, es_pr))
        print('{}: {}, lat: {}, lon: {}, cont: {}'.format(i, n, lat, lon, es_pr.get('Страна', '--')))
    else:
        print('{}: {}, TABLE IS NOT FOUND! link: {}'.format(i, n, lnk)) 
```
