```python
import urllib
import urllib.parse
import urllib.request
from bs4 import BeautifulSoup
```

```python
url_wikipedia = 'https://ru.wikipedia.org'
url_cities = 'https://ru.wikipedia.org/wiki/Категория:Города_по_субъектам_Российской_Федерации'
url_cities_q  = urllib.parse.quote(url_cities, safe='/:')
cities_bs = BeautifulSoup(urllib.request.urlopen(url_cities_q), 'lxml')
cities_bs
```

```python
all_cities={}
for links in cities_bs.find_all('div', class_='CategoryTreeItem'):
        for l in links.find_all('a'):
            all_cities[(l['title'], url_wikipedia+l['href'])] = {}
```

```python
def name_index_finder(cities_html):
    target_list = [item.text for item in cities_html[0]]
    targ_names = [
        "название",
        "Название",
        "город",
        "населённый пункт",
        "русское название",
        "название на русском",
        "названиерусское"
    ]
    for i,text in enumerate(target_list):
        if text in targ_names:
            return i
    else:
        return 1
```

```python
def special_finder(category_html):
    for subcategory in category_html.find_all('div', class_='mw-category-group'):
        for link in subcategory.find_all('a'):
            if 'Городские населённые пункты' in link['title']:
                target = url_wikipedia+link['href']
                return target
    else:
        return None
```

```python
all_cities    
```

```python
for category in list(all_cities.keys()):
    category_html = BeautifulSoup(urllib.request.urlopen(category[1]), 'lxml')
    target = special_finder(category_html)
    if not target:
        all_cities.pop(category)
        continue
    target_html = BeautifulSoup(urllib.request.urlopen(target), 'lxml')
    cities_html = target_html.find('table', class_='standard sortable').find_all('tr')
    try:
        name_index = name_index_finder(cities_html)
    except:
        print([item.text for item in cities_html[0].find_all('th')])
        print(target)
    for city in cities_html[1:]:
        city_a = city.find_all('td')[name_index].find('a')
        try:
            all_cities[category][(city_a['title'], url_wikipedia+city_a['href'])] = {}
        except:
            print(city_a)
```

```python
for h2 in city_html.find_all('h2'):
    for span in h2.find_all('span'):
        print(span)
```

```python
for cities in all_cities.values():
    for city, params in cities.items():
        city_html = BeautifulSoup(urllib.request.urlopen(city[1]), 'lxml')
        params['name'] = city_html.find('span', class_='mw-page-title-main').text
        for tr in city_html.find_all('table')[0].find_all('tr'):
            th = tr.find('th')
            td = tr.find('td')
            if th and td:
                params[th.text.strip()] = td.text.strip()
```
